# Get humidity data from 2002-2022

```python
import numpy as np
import pandas as pd
```

```python
# Đọc dữ liệu từ file Excel
file_path = 'V01.12.xlsx'  # Thay thế bằng đường dẫn thực tế đến file Excel của bạn
df = pd.read_excel(file_path)
```

```python
df.head()
```

### Replace the values .. to NaN

```python
df.replace("..", np.nan, inplace=True)
df.head(30)
```

### Add a header and replace the NaN years with the Year of the previous row

```python
df.columns = ['Year', 'Location'] + [str(i) for i in range(1, 13)]

# Chuyển các cột nhiệt độ từ kiểu object sang kiểu số
df.iloc[:, 2:] = df.iloc[:, 2:].astype(float)

# Thay thế các giá trị NaN trong năm dựa vào giá trị hàng trước đó
df['Year'].fillna(method='ffill', inplace=True)

print(df[:30])
```

### Find the average monthly humidity value of localities in 1 year

```python
avg_temp_by_month_year = df.groupby('Year')[df.columns[2:]].mean()
avg_temp_by_month_year.head(30)
```

```python
# Chuyển bảng thành dạng bạn mong muốn
stacked_df = avg_temp_by_month_year.stack().reset_index()

# Đặt lại tên cột
stacked_df.columns = ['Year', 'Month', 'Humidity']

# Xuất bảng đã chuyển đổi
print(stacked_df)
```

# Export average humidity data of Vietnam from 2002-2022

```python
stacked_df.to_csv('vietnam-humidity-handled-data.csv', index=False)
```
