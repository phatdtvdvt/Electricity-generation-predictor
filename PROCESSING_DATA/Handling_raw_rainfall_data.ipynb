# Get rainfall data from 1998-2015

```python
import pandas as pd

# Đọc dữ liệu từ file Excel
file_path = 'vietnam-rainfall-from-1901-2015-wb.xls'  # Thay thế bằng đường dẫn thực tế đến file Excel của bạn
df = pd.read_excel(file_path)

```

```python

df.head()
```

### remove tab in Year data column

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

### remove the Country, ISO3, ISO2 columns

```python
df = df.drop(columns=['Country', 'ISO3', 'ISO2'])
```

### Rearrange order

```python
new_order = ['Year', 'Month', 'pr']
df = df[new_order]
df = df.rename(columns={'pr': 'Rainfall'})
df = df[df['Year'] >= 1998]
df.head()
```

# Get temperature data from 2015 - 2022

```python
import numpy as np
```

```python
file_path2 = 'V01.09.xlsx'  # Thay thế bằng đường dẫn thực tế đến file Excel của bạn
df2 = pd.read_excel(file_path2)
```

```python
df2.head(30)
```

### Replace the values .. to NaN

```python
df2.replace("..", np.nan, inplace=True)
df2.head(30)
```

### Add a header and replace the NaN years with the Year of the previous row

```python
df2.columns = ['Year', 'Location'] + [str(i) for i in range(1, 13)]

# Chuyển các cột nhiệt độ từ kiểu object sang kiểu số
df2.iloc[:, 2:] = df2.iloc[:, 2:].astype(float)

# Thay thế các giá trị NaN trong năm dựa vào giá trị hàng trước đó
df2['Year'].fillna(method='ffill', inplace=True)

print(df2[:30])
```

### Find the average monthly rainfall value of localities in 1 year

```python
avg_temp_by_month_year = df2.groupby('Year')[df2.columns[2:]].mean()
avg_temp_by_month_year.head(30)
```

### Combine 2 data to 1998-2022

```python
# Chuyển bảng thành dạng bạn mong muốn
stacked_df = avg_temp_by_month_year.stack().reset_index()

# Đặt lại tên cột
stacked_df.columns = ['Year', 'Month', 'Rainfall']

# Xuất bảng đã chuyển đổi
print(stacked_df)
```

```python
merged_df = pd.concat([df, stacked_df], ignore_index=True)

merged_df.head(30)
```

# Export average rainfall data of Vietnam from 1998-2022

```python
merged_df.to_csv('vietnam-rainfall-handled-data.csv', index=False)
```
